```python
import numpy as np
import pandas as pd
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt 
```

```python
def generate_random_matrix(rows, cols):
    return np.random.rand(rows, cols)
def expmzbjb(A,q):
    n=A.shape[0]
    m = 0 if 0 > 1 + np.ceil(np.log2(np.linalg.norm(A,np.inf))) else 1 + np.ceil(np.log2(np.linalg.norm(A, 2)))
    A=A*2**(-m)
    D=np.eye(n)
    N=D
    X=D
    a=1
    for k in range(1,q+1):
        a=a*(q-k+1)/((2*q-k+1)*k)
        X=A.dot(X)
        N=N+a*X
        D=D+(-1)**k*a*X
    F=np.linalg.solve(D,N)
    for k in range(1,round(m+1)):
        F=F.dot(F)
    return F
```

```python
R=[]
for i in range(2,41):
    r=generate_random_matrix(i, i)
    R.append(r)
```

```python
expmsolve=[expm(r) for r in R]
expmjzbjbsolve3=[expmzbjb(r,3) for r in R]
expmjzbjbsolve4=[expmzbjb(r,4) for r in R]
expmjzbjbsolve5=[expmzbjb(r,5) for r in R]
expmjzbjbsolve6=[expmzbjb(r,6) for r in R]
expmjzbjbsolve9=[expmzbjb(r,9) for r in R]
expmjzbjbsolve12=[expmzbjb(r,12) for r in R]
error3=[]
error4=[]
error5=[]
error6=[]
error9=[]
error12=[]
for i in range(39):
    error3.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve3[i]))
    error4.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve4[i]))
    error5.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve5[i]))
    error6.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve6[i]))
    error9.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve9[i]))
    error12.append(np.linalg.norm(expmsolve[i]-expmjzbjbsolve12[i]))
```

```python
plt.plot(np.linspace(2,40,39),error3,label='q=3')
plt.xlabel('n-dimention of the matrix')
plt.ylabel('error(2-norm)')
plt.legend()
```
