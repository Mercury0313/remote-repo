```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import griddata

# 示例数据点
x = np.random.rand(100) * 4 - 2  # 生成-2到2之间的100个随机点
y = np.random.rand(100) * 4 - 2
values = np.zeros_like(x)  # 创建一个与x相同大小的零数组

# 创建网格
xi = np.linspace(-2, 2, 100)
yi = np.linspace(-2, 2, 100)
xi, yi = np.meshgrid(xi, yi)  # 创建网格点

# 插值
zi = griddata((x, y), values, (xi, yi), method='nearest')

# 绘制点密度图
plt.contourf(xi, yi, zi, cmap='Blues')
plt.colorbar(label='密度')

# 显示原始数据点
plt.scatter(x, y, marker='.', color='red')

# 显示图形
plt.title('二维点密度图示例')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.show()
```
