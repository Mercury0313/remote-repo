```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
facu=lambda x,t:1+np.sin(2*np.pi*(x+2*t))
```

```python
# 
h = 0.05  # 
tau = 0.02 # 
x_max = 1.0  # 
t_max = 0.8  # 
x_points = np.arange(0, x_max + h, h)  # 
t_points = np.arange(0, t_max + tau, tau)  # 
uacu1=[facu(i,0.8) for i in x_points] 
# 
u = 1+np.sin(np.pi*2*x_points)
# L-Fa=-2)a
U=[]
r=-2*tau/h
for t in t_points:
    U.append(u)
    u_new = np.zeros_like(u)
    for i in range(1,len(u)-1):
        u_new[-1]=1+np.sin(np.pi*4*(t+tau))
        u_new[0]=u[0]+2*tau/h*(u[1]-u[0])#
        u_new[i] = (1-r)/2*(u[i+1])+(1+r)/2*u[i-1]
    u = u_new
plt.plot(x_points,U[-1],label='L-F solution')
plt.plot(x_points,uacu1,label='accurate solution')
plt.xlabel('x')
plt.ylabel('u')
plt.title('solution(t=0.8)') 
plt.legend()
ua=U[-1]
error=[np.linalg.norm(U[-1][i]-uacu1[i]) for i in list(range(21))]
plt.figure()
plt.plot(x_points,error,label='error')
plt.legend()
ea=error
```
